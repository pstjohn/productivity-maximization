```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
sns.set_style('darkgrid')
sns.set_context('talk', font_scale=1.5)
sns.set(color_codes=True)

%matplotlib inline

def lighten_color(color, degree):
    cin = matplotlib.colors.colorConverter.to_rgb(color)
    cw = np.array([1.0]*3)
    return tuple(cin + (cw - cin)*degree)
```

```python
from scipy.interpolate import lagrange

def interp_constraints(coll, constraint, ts):
    const_array = np.array([[coll._constraint_fns[constraint](
                    {met : var_op for met, var_op in
                     zip(coll.boundary_species, coll.var.x_op[ki,ji])})[0]
                             for ji in range(0, coll.d+1)]
                            for ki in range(coll.nk)])
    
    out = np.empty(len(ts))
    
    stage_starts = [0.]
    for i in range(coll.nk-1):
        stage_starts += [coll.var.h_op[coll._get_stage_index(i)] +
                         stage_starts[-1]]
    stage_starts = pd.Series(stage_starts)
    stages = stage_starts.searchsorted(ts, side='right') - 1

    for ki in range(coll.nk):
        interp = lagrange(coll.col_vars['tau_root'], 
                          const_array[ki])

        out[stages == ki] = interp(
            (ts[stages == ki] - stage_starts[ki]) /
            coll.var.h_op[coll._get_stage_index(ki)])
        
    return out
```

```python
#stages = [20]*1
#stages = [10]*2
stages = [5]*4
```

```python
from ecoli.coll_setup_maxod import (setup_collocation, initialize_collocation,
                                     calc_max_productivity, calc_max_yield)

ecoli_coll = setup_collocation(stages)
calc_max_productivity(ecoli_coll)
coll = ecoli_coll

ts = np.linspace(0, coll.var.tf_op, 100)
sol = coll._interpolate_solution(ts)
deriv = coll._interpolate_derivative(ts)

glc_ind = coll.boundary_species.index('glc__D_e')
o2_ind = coll.boundary_species.index('o2_e')

fstage_starts = np.hstack([np.array(0.),
                           np.cumsum(np.array(coll.stage_breakdown) *
                                     coll.var.h_op)])

fstage_ends = np.hstack([np.cumsum(np.array(coll.stage_breakdown) *
                                   coll.var.h_op), coll.var.tf_op])
```

```python
with sns.axes_style('ticks'):
    fig = plt.figure(figsize=(8,4))

    import matplotlib.gridspec as gridspec
    gs = gridspec.GridSpec(2, 3)
    ax1 = fig.add_subplot(gs[0, :])

    ax_biomass = ax1.twinx()

    ax2 = fig.add_subplot(gs[1, 0], sharex=ax1)
    ax3 = fig.add_subplot(gs[1, 1], sharex=ax1)
    ax4 = fig.add_subplot(gs[1, 2], sharex=ax1)

    conc_lines = ax1.plot(ts, sol[:,-4:-1])
    biomass_lines = ax_biomass.plot(ts, sol[:,0], color=sns.color_palette()[-2])

    ax2.plot(ts, deriv[:,1] / sol[:,0])
    ax2.plot(ts, interp_constraints(coll, 'atpm', ts), 'r--')

    ax3.plot(ts, -deriv[:, glc_ind] / sol[:,0], label='Optimal flux')
    ax3.plot(ts, -interp_constraints(coll, 'glc__D_e', ts), 'r--', label='Constraint')
    
    ax4.plot(ts, -deriv[:, o2_ind] / sol[:,0], label='Optimal flux')
    ax4.plot(ts, -interp_constraints(coll, 'o2_e', ts), 'r--', label='Constraint')

    ax1.set_xlabel('Time (hr)')
    ax3.set_xlabel('Time (hr)')

    ax2.set_title('ATP Maintenance')
    ax3.set_title('Glucose Uptake')
    ax4.set_title('Oxygen Uptake')
    
    ax2.set_ylabel('mM / g DCW * hr')
    
    ax_biomass.set_ylabel('Biomass (g DCW/L)')
    ax1.set_ylabel('Concentration (mM)')
    
    ax_biomass.set_ylim([0, np.ceil(sol[:,0].max())])
    
    #ax1.set_xlim([0, coll.var.tf_op])
    
    sns.despine(ax=ax1, right=False, top=True)
    sns.despine(ax=ax_biomass, right=False, top=True)
    sns.despine(ax=ax2)
    sns.despine(ax=ax3)
    sns.despine(ax=ax4)

    ax_biomass.locator_params(tight=True, nbins=6)
    ax4.locator_params(tight=True, nbins=6)
    ax1.locator_params(tight=True, nbins=6)

    
    ax1.legend(conc_lines + biomass_lines,
               ['Glucose', 'Succinate', 'Acetate', 'Biomass'],
              loc='center left')
    
    ax3.legend(loc='lower left')
    
    fig.subplots_adjust(hspace=0.75, wspace=0.3)
    
    colors = sns.color_palette("Set2", len(coll.stage_breakdown)) 
    for ax in [ax1, ax2, ax3, ax4]:
        for start, end, color in zip(fstage_starts, fstage_ends, colors):
            ax.axvspan(start, end, facecolor=lighten_color(color, 0.85), edgecolor='none')
    
    ax2.set_ylim([8, 30])
    ax3.set_ylim([0, 10])
    ax4.set_ylim([0, 20])
    ax1.set_xlim([0, 15.325])
    ax1.set_ylim([0, 450])

    ax_biomass.set_ylim([0, 16])
    
    ax1.set_title('E. coli maximum succinate producitivity')
    fig.savefig('figs/ecoli{0:02d}stage.svg'.format(len(coll.stage_breakdown)))
```

```python
from succinogenes.coll_setup import (setup_collocation, initialize_collocation,
                                     calc_max_productivity, calc_max_yield)

succ_coll = setup_collocation(stages)
calc_max_productivity(succ_coll)
coll = succ_coll

ts = np.linspace(0, coll.var.tf_op, 100)
sol = coll._interpolate_solution(ts)
deriv = coll._interpolate_derivative(ts)

glc_ind = coll.boundary_species.index('glc__D_e')

fstage_starts = np.hstack([np.array(0.),
                           np.cumsum(np.array(coll.stage_breakdown) *
                                     coll.var.h_op)])

fstage_ends = np.hstack([np.cumsum(np.array(coll.stage_breakdown) *
                                   coll.var.h_op), coll.var.tf_op])

with sns.axes_style('ticks'):
    fig = plt.figure(figsize=(8,4))

    import matplotlib.gridspec as gridspec
    gs = gridspec.GridSpec(2, 2)
    ax1 = fig.add_subplot(gs[0, :])

    ax_biomass = ax1.twinx()

    ax2 = fig.add_subplot(gs[1, 0], sharex=ax1)
    ax3 = fig.add_subplot(gs[1, 1], sharex=ax1)
  
    conc_lines = ax1.plot(ts, sol[:,-4:-1])
    biomass_lines = ax_biomass.plot(ts, sol[:,0], color=sns.color_palette()[-2])

    ax2.plot(ts, deriv[:,1] / sol[:,0])
    ax2.plot(ts, interp_constraints(coll, 'atpm', ts), 'r--')

    ax3.plot(ts, -deriv[:, glc_ind] / sol[:,0], label='Optimal flux')
    ax3.plot(ts, -interp_constraints(coll, 'glc__D_e', ts), 'r--', label='Constraint')
    
 
    ax1.set_xlabel('Time (hr)')
    ax2.set_xlabel('Time (hr)')
    ax3.set_xlabel('Time (hr)')

    ax2.set_title('ATP Maintenance')
    ax3.set_title('Glucose Uptake')
    
    ax2.set_ylabel('mM / g DCW * hr')
    
    ax_biomass.set_ylabel('Biomass (g DCW/L)')
    ax1.set_ylabel('Concentration (mM)')
    
    ax_biomass.set_ylim([0, np.ceil(sol[:,0].max())])
    
    #ax1.set_xlim([0, coll.var.tf_op])
    
    sns.despine(ax=ax1, right=False, top=True)
    sns.despine(ax=ax_biomass, right=False, top=True)
    sns.despine(ax=ax2)
    sns.despine(ax=ax3)

    ax_biomass.locator_params(tight=True, nbins=6)
    ax1.locator_params(tight=True, nbins=6)

    
    ax1.legend(conc_lines + biomass_lines,
               ['Glucose', 'Succinate', 'Acetate', 'Biomass'],
              loc='center left')
    
    ax3.legend(loc='lower left')
    
    fig.subplots_adjust(hspace=0.75, wspace=0.3)
    
    colors = sns.color_palette("Set2", len(coll.stage_breakdown)) 
    for ax in [ax1, ax2, ax3]:
        for start, end, color in zip(fstage_starts, fstage_ends, colors):
            ax.axvspan(start, end, facecolor=lighten_color(color, 0.85), edgecolor='none')
    
    ax2.set_ylim([4, 11])
    ax3.set_ylim([0, 10])
    ax1.set_xlim([0, 19.26])
    ax1.set_ylim([0, 250])
    ax_biomass.set_ylim([0, 5])
    
    ax1.set_title('A. succinogenes maximum succinate producitivity')
    fig.savefig('figs/succ{0:02d}stage.svg'.format(len(coll.stage_breakdown)))
```
